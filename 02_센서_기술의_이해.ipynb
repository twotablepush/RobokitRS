# LiDAR

- "Light Detection and Ranging"의 약자로, 광 센싱 기술 중 하나이다. 크게 레이저 송신부, 수신부, 데이터 처리부로 구성되어 있으며 주로 거리 측정에 사용된다. 일련의 빠른 레이저 광원를 방출하여 목표물에서 튕겨 되돌아 오는 레이저의 속도를 계산하여 물체의 거리를 측정한다.

- 일반적으로 빠른 펄스 반복률의 레이저, 스캐너, GPS 수신기로 구성. 이 요소들은 함께 작동하여 물체의 위치와 거리를 정확하게 측정하고 그 결과를 높은 해상도의 3D지도로 변환한다. 지도는 건물, 나무, 도로 그리고 다른 피쳐들의 위치와 모양을 매우 정확하게 나태낼 수 있다.

- 매우 많은 분야에서 활용된다. 지질학자들은 LiDAR를 사용하여 지형을 매핑하고, 홍수 위험 지역을 식별하며, 화산의 활동을 모니터링한다. 또한 생물학자들은 이 기술을 사용하여 숲의 구조를 분석하고, 동물들의 서식지를 이해하는데 도움을 받는다.

- 자율주행 자동차에서는 물체와 장애물을 감지하고 위치를 파악하는 데 사용되며 이를 통해 차량이 안전하게 운행될 수 있도록 돕는다.

- 그러나 비용이 높고 부피가 크다는 단점이 있다. 이로 인해 일부 회사는 카메라나 레이더 기반의 시스템으로 대체하고 있다. 그럼에도 불구하고 많은 자율주행 차량 개발 회사는 LiDAR의 높은 정밀도와 신뢰성 때문에 여전히 선호하고 있다.

- 파장이 짧은 레이저를 사용하므로 레이더보다 측정 정밀도와 공간 해상도가 높아, 물체 크기, 형태를 빠르게 파악할 수 있으며 레이저 빔을 광 시야각으로 확장할 경우 3D Scanning도 가능하다.

- 미세먼지 크기까지도 측정이 가능하기 때문에 레이저가 이동하는 동안 일정 크기 이하의 물체는 필터링이 가능하다.

- 레이저 센서에 비해 감지 범위가 넓고 감지 반응 속도가 높다.

- 레이저가 여러 Layer로 방출되어 측정되고, 대개의 경우 각 Layer별로 감지 설정도 가능하며 입체적으로 물체의 형상도 확인 가능하다.

- 날씨, 계절, 태양광의 영향을 받으므로 설치방법에 따라 환경의 영향을 받는다.



# RADAR

- "Radio Detection and Ranging"의 약자로서, 전파를 이용하여 물체를 탐지하고 거리를 측정하는 기술이다. 광학적인 센서인 LiDAR나 카메라와 달리 악천후 환겨에서도 잘 작동하며, 물체의 속도를 직접 측정할 수 있다.

- 자율주행 차량에서 RADAR
 - 물체 탐지 : 주변 환경에서 물체를 탐지하고 그 물체까지의 거리와 속도를 측정하는 데 사용된다. 이는 주행 경로에 장애물이 있는지, 앞서 가는 차량이 있는지 등을 파악하는 데 유용하다.

 - 속도 측정 : 도플러 효과를 이용하여 물체의 상대적인 속도를 측정한다. 이 기능은 차량이 앞서가는 차량의 속도를 추정하거나, 스스로의 속도를 조절하는 데 사용된다.

 - 악천후 환경에서의 성능 : 안개, 비, 눈 등의 악천후 환경에서도 잘 작동한다. 빛을 기반으로 하는 센서인 LiDARR나 카메라가 이러한 조건에서 성능 저하를 겪는 것과 대조적이다.

- RADAR는 비교적 저렴하고 내구성이 있어 많은 자율주행 차량에서 채택되고 있다. 그러나 해상도가 비교적 낮아 물체릐 세부적인 형태를 인식하는 데는 한계가 있다. 따라서 많은 자율주행 차량은 RADAR, LiDAR, 카메라 등을 결합하여 각 센서의 장단점을 보완한다.





# 컴퓨터비전

- 컴퓨터가 디지털 이미지나 비디오를 인식하고 이해하는 기술이다. 이는 인간의 시각 체계를 모방하여 컴퓨터가 물체, 사람, 텍스트, 동작등을 인식하고 분석하도록 한다.

- 컴퓨터비전의 역할
 - 환경 인식 : 차량의 주변 환경을 이해하는데 도움이 된다. 카메라를 통해 캡처된 이미지를 분석하여 차량, 보행자, 도로 표시, 교통 신호 등을 인식한다.

 - 경로 계획 : 차량의 경로를 계획하는데 사용된다. 차선을 인식하여 차량이 올바른 차선을 유지할 수 있도록 돕거나, 앞에 장애물이 있을 경우 이를 피할 수 있는 경로를 계획한다.

 - 사람과의 상호작용 : 차량이 사람의 행동을 이해하고 예측하는 데 중요하다. 보행자의 움직임을 인식하여 그들의 의도를 예측하거나, 다른 운전자의 신호를 해석할 수 있다.

- 머신러닝, 특히 딥러닝과 밀접하게 연관되어 있다. 딥러닝 알고리즘은 대량의 이미지 데이터를 통해 학습하여 물체를 인식하고 분류하는 능력을 향상시킨다. 이러한 기술은 자율주행 차량의 발전에 매우 중요한 역할을 하고 있다.

- 그러나 컴퓨터비전도 완벽하지는 않다. 악천후나 낮은 조명 조건에서는 성능이 저하될 수 있으며, 복잡하거나 예기치 않은 상황에 대응하는 능력이 제한적일 수 있다. 따라서 컴퓨터비전은 RADAR나 LiDAR와 같은 다른 센서 기술과 결합되어 사용되는 경우가 많다.
