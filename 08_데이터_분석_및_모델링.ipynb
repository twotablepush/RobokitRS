# Radar
- 레이더 센서는 전자기파를 생성하여 주변 물체와의 거리, 속도, 방향을 측정하는 데 사용.
- 주로 전방 및 후방에 설치되어 차량 간 거리 유지, 충돌 회피, 차선 변경 보조 등에 활용된다.
- 레이터 센소는 고주파 전자기파를 발생시켜 반사되는 신호를 수신하여 객체의 위치와 속도를 측정.
- 데이터 분석 및 모델링은 수신된 레이더 데이터를 처리하여 주변 물체의 위치, 속도, 이동 경로 등을 추정하고 이를 기반으로 운전 상황을 판단.

# Camera
- 카메라 센서는 영상을 캡처하여 주변 환경을 시각적으로 인식하는 데 사용된다.
- 주로 전방, 후방, 측면 등에 설치되어 도로 상황, 표지판, 신호등, 차선 등을 인식.
- 컴퓨터 비전 기술과 결합하여 객체 검출, 객체 분류, 차선 인식, 신호등 인식 등의 작업을 수행.
- 데이터 분석 및 모델링은 카메라로부터 캡처된 영상 데이터를 처리하여 객체의 위치, 크기, 모양 등을 추출하고 이를 기반으로 운전 상황을 분석.

# Ultrasonic
- 초음파 센서는 초음파를 사용하여 주변 장애물과의 거리를 감지하는 데 사용된다.
- 일반적으로 차량 주변에 설치되어 주차 시에 사용되며, 주변 환경의 근접 여부를 측정.
- 초음파 센서는 초음파 신호를 발생시켜 반사되는 신호를 수신하여 거리를 측정.
- 데이터 분석 및 모델링은 초음파 센서로 부터 수집된 거리 데이터를 처리하여 주변 장애물과의 거리, 방향 등을 분석한다. 주로 주차 보조 시스템이나 충돌 회피 시스템에 활용.

# Lidar
- 라이다 센서는 레이저 빛을 사용하여 주변 환경을 3D로 스캔하고 거리를 측정하는 데 사용된다.
- 레이저 빛을 발사하여 주변 물체에 반사된 신호를 수신하여 거리, 위치, 크기 등의 정보를 파악.
- 라이다 센서는 높은 정밀도와 고해상도를 제공하며, 자율 주행 자동차에서 장애물 감지, 환경 맵핑, 경로 계획 등에 활용된다.
- 데이터 분석 및 모델링은 라이다 센서에서 수집된 3D 포인트 클라우드 데이터를 처리하여 객체의 위치, 크기, 모양 등을 추출하고, 환경 맵핑이나 장애물 감지에 활용.

# Gyroscope
- 자이로스코프 센서는 차량의 회전 속도와 방향을 감지하는 데 사용.
- 차량의 기울기, 회전 등의 동작을 측정하여 운전 상황을 파악.
- 주로 자동차의 방향 제어, 차체 안정성 제어 등에 활용.
- 데이터 분석 및 모델링은 자이로스코프 센서에서 수집된 회전 데이터를 처리하여 차량의 움직임, 기울기 등을 추적하고, 운전 상황에 대한 정보를 분석한다.
