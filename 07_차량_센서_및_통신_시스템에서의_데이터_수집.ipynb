# 모빌리티 센서 데이터 수집
- GPS(Global Positioning System) : GPS 수신기를 사용하여 위성 신호를 수신하고, 이를 통해 차량의 위치 데이터를 수집.
- 가속도계(Accelerometer) : 차량의 가속도를 측정하는 가속도계를 사용하여 가속도 데이터를 수집.
- 자이로스코프(Gyroscope) : 차량의 회전 속도와 방향을 측정하는 자이로스코프를 사용하여 회전 관련 데이터를 수집.
- 라이다(LIDAR, Light Detection and Ranging) : 레이저 빔을 사용하여 주변 환경의 거리와 형상을 수집한다.
- 카메라(Camera) : 차량 주변의 시각 정보를 캡처하는 카메라를 사용하여 이미지 또는 비디오 데이터를 수집.

GPS/IMU : IMU는 광성 센서를 의미한다. 관성 센서는 엑셀로미터나 자이로와 같은 센서를 통칭하는 말.
GPS는 GNSS 통신 방식이 최근 유행이다. 여러 개의 GPS위성으로부터 위치 신호를 수신할 수 있는 체계이다. 기존 GPS는 위성 3개를 통해 자신의 위치를 파악했으나 GNSS는 그 보다 더 많은 GPS 위성으로부터 위치 정보를 수신하여 위치를 파악할 수 있다.

# 차량 내 데이터 저장
- 센서로부터 수집된 데이터는 차량 내의 저장 장치에 저장된다.
- 일반적으로 플래시 메모리 또는 하드 드라이브와 같은 저장 장치를 사용한다.
- 저장된 데이터는 차량 내 분석, 운전 기록, 사고 조사 등에 활용.

# 외부 서버 또는 클라우드 저장
- 센서 데이터는 안전한 인터넷 연결을 통해 외부 서버 또는 클라우드에 저장될 수 있다. 이를 통해 데이터는 중앙 집중화되고, 보다 큰 규모의 분석 및 응용 프로그램에 활용될 수 있다. 클라우드 저장소는 확장성과 가용성을 제공하여 대량의 데이터를 관리할 수 있다.
